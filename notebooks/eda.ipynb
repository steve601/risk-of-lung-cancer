```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### let's load the clean data

```python
df = pd.read_csv('C:/Users/odhia/OneDrive/Desktop/datascienceproject/data/clean_df.csv')
df.head()
```

### distribution plots for the columns

```python
cat_cols = ['gender','smoking', 'yellow_fingers', 'anxiety',
       'peer_pressure', 'chronic_disease', 'fatigue', 'allergy', 'wheezing',
       'alcohol_consuming', 'coughing', 'shortness_of_breath',
       'swallowing_difficulty', 'chest_pain']
for col in cat_cols:
    print(df[col].value_counts())
    print('='*10)
```

```python
for i in cat_cols:
    sns.countplot(x=df[i])
    plt.title(f'Count distribution of {i}')
    plt.show()
```

### plot for mean_smoothness against mean_radius,with respect to diagnosis

```python
for i in cat_cols:
    sns.countplot(x=df[i],hue=df['lung_cancer'])
    plt.title(f'Count distribution of {i} with respect to Lung cancer')
    plt.show()
```
